## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
QUERY = """
SELECT ConvertedComp
FROM master
"""

df = pd.read_sql_query(QUERY,conn)
df.hist(bins = 10)
plt.title('Histogram of ConvertedComp')
plt.xlabel('ConvertedComp')
plt.ylabel('Density')
```

### Box Plots


Plot a box plot of `Age.`


```python
QUERY = """
SELECT Age
FROM master
"""

df = pd.read_sql_query(QUERY,conn)
df.boxplot()
plt.title('Boxplot of Age')
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
QUERY = """
SELECT Age, WorkWeekHrs
FROM master
"""

df = pd.read_sql_query(QUERY,conn)
df.plot(kind = 'scatter', x ='Age', y = 'WorkWeekHrs' )
plt.title('Scatter pot of Age and WorkWeekHrs')
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
"""

df = pd.read_sql_query(QUERY,conn)

# Scale the Age values to make bubble sizes visually distinct
bubble_size = df['Age'] / df['Age'].max() * 1000  # Adjust the scaling factor if needed

df.plot(kind = 'scatter', x = 'WorkWeekHrs', y = 'CodeRevHrs', s = bubble_size, alpha = 0.5)
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
QUERY = """
SELECT DatabaseDesireNextYear, COUNT (*) as Count
FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY Count DESC LIMIT 5

"""

df = pd.read_sql_query(QUERY,conn)

df.set_index(df['DatabaseDesireNextYear'], inplace = True)

Size = df['Count']
Labels = df.index

plt.pie(Size, labels = Labels, autopct = '%1.1f%%', startangle = 90)
plt.title('Top 5 databases that respondents wish to learn next year')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
WHERE Age BETWEEN 30 AND 35
"""

df = pd.read_sql_query(QUERY,conn)

# Calculate the median values for WorkWeekHrs and CodeRevHrs
median_workweekhrs = df['WorkWeekHrs'].median()
median_coderevhrs = df['CodeRevHrs'].median()

# Prepare data for the stacked bar chart
data = pd.DataFrame({'WorkWeekHrs': [median_workweekhrs],'CodeRevHrs': [median_coderevhrs]}, index=['Median Hours'])

data.plot(kind = 'bar', alpha = 0.25, stacked = True)
plt.title('median WorkWeekHrs and CodeRevHrs for the age group 30 to 35')
plt.ylabel('Hours')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
QUERY ="""
SELECT ConvertedComp, Age
FROM master
WHERE Age BETWEEN 45 AND 60
"""

df = pd.read_sql_query(QUERY,conn)

df = df.groupby('Age')['ConvertedComp'].median()

df.plot(kind = 'line')
plt.title('median ConvertedComp for all ages from 45 to 60')
plt.ylabel('median converted comp')
plt.show()

```
